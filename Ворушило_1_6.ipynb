```python
# Завдання 1 Створіть одновимірний масив (вектор) з перших 10-ти додатних парних чисел та виведіть його значення.

array = [i for i in range(2,11,2)]
print("Масив :", array)
```

```python
# Завдання 2 Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його сумою кожного числа та його попереднього, починаючи з 1, та виведіть його значення.

v = 1
matrix = []
for i in range(3):
    row = []
    for j in range(3):
        nv = v + (v - 1)
        row.append(nv)
        v += 1
    matrix.append(row)
for row in matrix:
    print(row)
```

```python
# Завдання 3 Створіть масив розміром 2x3, заповніть його випадковими цілими числами в діапазоні від -1 до 1 та виведіть його значення.

matrix1 = np.random.randint(-1, 1, size =(2,3))
print(matrix1)
```

```python
# Завдання 4 Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

matrix2 = np.random.random( size =(5,5))
print(matrix2)
row_sums = matrix2.sum(axis = 1)
print("\n", row_sums.reshape(-1,1))
```

```python
# Завдання 5 Створіть два одновимірних масиви розміром 5, заповніть їх випадковими дійсними числами від 0 до 1 та виконайте на них поелементні операції додавання, віднімання та множення.

b = np.random.random( size = 5)
c = np.random.random( size = 5)
print("b = ", b, "c = ", c)

sum = b + c
sub = b - c
mult = b * c
print("+ = ", sum, "\n- =", sub, "\n* = ", mult)
```

```python
# Завдання 6 Створіть дві матриці 3x3, заповніть їх довільними дійсними числами від 5 до 30 та знайдіть їх добуток.

matrix1 = np.random.randint(5, 30, size = (3, 3))
matrix2 = np.random.randint(5, 31, size = (3, 3))
print("1 матриця", "\n", matrix1, "\n2 матриця", "\n", matrix2)
dob = np.dot(matrix1, matrix2)
print("добуток матриць = ", "\n", dob)
```

```python
# Завдання 7 Створіть дві матриці розміром 2x3 та 3x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

matrix1 = np.random.randint(1, 6, size = (2, 3))
matrix2 = np.random.randint(1, 6, size = (3, 4))
print("1 матриця", "\n", matrix1, "\n2 матриця", "\n", matrix2)
dob1 = np.dot(matrix1, matrix2)
print("добуток матриць = ", "\n", dob1)
```

```python
# Завдання 8 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 8 та знайдіть її обернену матрицю.

matrix = np.random.randint(1, 9, size = (3, 3))
print("матриця", "\n", matrix)
matrix_inv = np.linalg.inv(matrix)
print("обернена матриця", "\n", matrix_inv)
```

```python
# Завдання 9 Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

matrix = np.random.random(size = (4, 4))
print("матриця", "\n", matrix)
print("транспонована матриця", "\n", matrix.T)
```
